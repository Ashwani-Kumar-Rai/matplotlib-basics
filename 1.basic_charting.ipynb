```python
from matplotlib import pyplot as plt 
from matplotlib import style # to use a style
style.use('ggplot')  # to use a particular style
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]

plt.plot(x,y)
plt.plot(x2,y2)

```

```python
from matplotlib import pyplot as plt 
from matplotlib import style # to use a style
style.use('ggplot')  # to use a particular style
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]

plt.plot(x,y ,label="first")
plt.plot(x2,y2,label="second")
plt.legend() # legend in in top right corner first , second
plt.show()

```

```python
from matplotlib import pyplot as plt 
from matplotlib import style # to use a style
style.use('ggplot')  # to use a particular style
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]

plt.plot(x,y ,label="first",linewidth=4) # linewidth=4 makes the line thick 
plt.plot(x2,y2,label="second")
plt.legend() # legend in in top right corner first , second
plt.show()

```

```python
from matplotlib import pyplot as plt 
from matplotlib import style # to use a style
style.use('ggplot')  # to use a particular style
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]

plt.scatter(x,y ,label="first",linewidth=4) # linewidth=4 makes the line thick 
plt.plot(x2,y2,label="second")
plt.legend() # legend in in top right corner first , second
plt.show()

```
